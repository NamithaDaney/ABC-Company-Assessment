**Discover which team and position have the highest salary expenditure.**

```python
highest_Salary = df.groupby(["Team",'Position'])['Salary'].sum()
high_Salary = highest_Salary.max()
position = highest_Salary.idxmax()
print(f"The team and position have the highest salary expenditure is {position} with salary of {high_Salary}")
```

**Investigate if there's any correlation between age and salary, and represent it visually.**

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.scatterplot(x=df['Age'], y=df['Salary'])
plt.title('Age vs Salary')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.show()
```

From this graph it is evident that is no specific correlation between age and salary. Employees accross all age groups appers to have almost similar minimum salary level.

```python

#Determine the distribution of employees across each team and calculate the percentage split relative to the total number of employees.
from matplotlib import figure
plt.figure(figsize=(20,10))
plt.pie(df['Team'].value_counts(), labels=df['Team'].value_counts().index, autopct='%1.1f%%')
plt.title('Distribution of employees across each team')
plt.show()
```

The distribution of employees across each team almost similar.The percentage distribution of most team is close to 3.3 percent. The team 'New Orleans Pelicans' has the highest percentage of employees(4.1per cent.)

```python
#Segregate employees based on their positions within the company
plt.figure(figsize=(10,5))
sns.countplot(x='Position', data=df)
plt.title('Distribution of Employees by Position')
plt.xlabel('Position')
plt.ylabel('Count')
plt.show()
```

The position 'SG' has highest number of employees(110). The position 'C' has comparatively lower number of employees(80).
